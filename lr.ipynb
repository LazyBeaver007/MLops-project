```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataa = pd.read_csv("Salary_Data.csv")

from sklearn.model_selection import train_test_split


dataa.head()
```

```python
X= dataa.drop("YearsExperience",axis=1)

Y = dataa.drop("Salary",axis=1)
print(X.head())
print(Y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42 )
```

```python
plt.scatter(X_train,y_train)
plt.xlabel("Exp")
plt.ylabel("salary")
plt.show()
```

```python
def cost_function(X,Y,w,b):
    m=len(X)
    cost_sum = 0

    for i in range(m):
        f = w * X[i]+b
        cost = (f-Y[i])**2
        cost_sum += cost 
    total_cost = (1/(2*m))*cost_sum
    return total_cost
```

```python
def grad(X,Y,w,b):
    m = len(X)
    dc_dw = 0
    dc_db = 0 
    for i in range(m):
        f = w*X[i]+b
        dc_dw +=(f-Y[i])*X[i]
        dc_db +=(f-Y[i])
    dc_db = (1/m)*dc_db
    dc_dw = (1/m)*dc_dw 
    return dc_dw,dc_db
```

```python
def gradd(x,y,alpha, iterations):
    w=0
    b=0

    for i in range(iterations):
        dc_dw,dc_db = grad(x,y,w,b)
        w = w-alpha*dc_dw
        b = b - alpha*dc_db
        print(f"iter{i}: cost{cost_function(x,y,w,b)}")


    return w,b 
```

```python
lr = 0.00001
iterr = 50

fw,fb = gradd(X_train,y_train,lr,iterr)
print(f"w:{fw:.2f},b:{fb:.2f}")
```

```python
# convert DataFrame columns to 1D numpy arrays for the gradient descent functions
X_train = X_train['Salary'].values
y_train = y_train['YearsExperience'].values
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot of training data
plt.scatter(X_train, y_train, label='Data Points')

# Generate x values for the regression line
x_vals = np.linspace(min(X_train), max(X_train), 100)

# Compute corresponding y values using final weights
y_vals = fw * x_vals + fb

# Plot the regression line
plt.plot(x_vals, y_vals, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
